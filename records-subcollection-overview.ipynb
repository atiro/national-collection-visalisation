```python
import pandas as pd
import altair as alt
records_data = pd.read_csv("records-sub-stats-all.csv")
```

# Records (with sub-collections)

(see below for explaination of why some terms are in quotes)

This is an attempt to highlight particular "subcollections" in institutions, to give more indication of where cataloguing may be more complete in one area than in another. It should be noted it *may be* but is not always an indication of different cataloguing types of records (i.e. museum, library, archive). Even if an institution has examples of all three, that not automatically mean three sub-collections. Equally, an institution might have three sub-collections, but this does not mean it contains a museum, library and archive, it might be for example three different distinct museum collections that make up the overall institution. 


## Collection with subcollections record size by topic

```python
title = alt.TitleParams('Collection Size (Records) by topic', anchor='middle')
alt.Chart(records_data,title=title).mark_bar().encode(
    alt.Color('institution:N').legend(orient="bottom", titleOrient="left", columns=4),
    x='record_count:Q',
    tooltip=['institution', 'subcollection', 'record_count'],
    y='topic:N',
    column='sector'
).properties(width=220).resolve_scale(x='independent').configure(numberFormat='.2s')
```

## Collection with subcollections record size by precision

```python
import altair as alt

title = alt.TitleParams('Collection Size (Artefact Records) by precision', anchor='middle')
alt.Chart(records_data, title=title).mark_bar().encode(
    column='precision:O',
    x='record_count:Q',
    y='topic:N',
    tooltip='subcollection',
    color=alt.Color('institution:N', sort='descending').legend(orient="bottom", titleOrient="left", columns=4)
).properties(width=220).resolve_scale(x='independent').configure(numberFormat='.2s')
```

## Collection with subcollections record size by precision

```python
import altair as alt

title = alt.TitleParams('Records per institution by named sub-collections', anchor='middle')
alt.Chart(records_data, title=title).mark_bar().encode(
        alt.Color('subcollection:N').legend(orient="bottom", columns=4, titleOrient="left"),
    column='precision:O',
    tooltip=['institution', 'subcollection', 'record_count'],
    x='record_count:Q',
    y='institution:N'
).properties(width=250).resolve_scale(x='independent').configure(numberFormat='.2s')
```
